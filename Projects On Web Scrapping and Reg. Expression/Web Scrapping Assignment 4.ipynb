1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank
B) Name
C) Artist
D) Upload date
E) Views 

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
driver.implicitly_wait(10)
```

```python
Name = []
Artist = []
Upload_date = []
Views = []
```

```python
name_ = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in name_:
    if i.text is None:
        Name.append("--") 
    else:
        Name.append(i.text)
    
art = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in art:
    if i.text is None:
        Artist.append("--")
    else:
        Artist.append(i.text)
        
upload = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in upload:
    Upload_date.append(i.text)

view = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in view:
    Views.append(i.text)
```

```python
len(Name),len(Artist),len(Upload_date),len(Views)
```

```python
df = pd.DataFrame({'Name':Name,'Artist':Artist,'Upload_date':Upload_date,'Views':Views})
df
```

```python
# USING LAMBDA FUNCTION TO REMOVE COLUMNWISE CHARACTERS BY USING -4 FROM LAST CHARACTER
df.Name = df.Name.apply(lambda x:x[0:-4].strip('"')) #  USING STRIP FUNCTION TO REMOVE INVERTED COMMA
df
```

```python
len(url)
```

```python
#Create lists
repo_title = []
repo_desc = []
c_counts = []
lang = []
```

```python
len(repo_title),len(repo_desc),len(c_counts),len(lang)
```

```python
df = pd.DataFrame({'Repositoty':repo_title,'Repository Description':repo_desc,'Contributors':c_counts,'Language':lang})
df
```

8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You
have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year 

```python
driver = webdriver.Chrome()
driver.maximize_window()
driver.get('https://archive.ics.uci.edu/')
driver.implicitly_wait(10)
```

```python
# click on view dataset
button = driver.find_element(By.XPATH,'//a[@class="btn-primary btn"]').click()
```

```python
#fetching url of each datasets
datasets_urls = []
start = 0
end = 1
for page in range(start,end):
    url = driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in url:
        datasets_urls.append(i.get_attribute('href'))
    next_button=driver.find_elements(By.XPATH,'//button[@class="btn-primary btn-sm btn"]')
```

```python
len(datasets_urls)
```

```python
datasets_urls
```

```python
Dataset_name = []
Data_type = [] 
Task = []
Attribute_type = []
No_of_instances = []
No_of_attribute = []
Year = []
```

```python
for i in datasets_urls:
    driver.get(i)
    try:
        name = driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        Dataset_name.append(name.text)
    except NoSuchElementException:
        Dataset_name.append('Not Present')

    try:
        type_of_data = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p')
        Data_type.append(type_of_data.text)
    except NoSuchElementException:
        Data_type.append('Not Present')

    try:
        task = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[2]/p')
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('Not Present')

    try:
        attribute_type = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
        Attribute_type.append(attribute_type.text)
    except NoSuchElementException:
        Attribute_type.append('Not Present')

    try:
        instances = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p')
        No_of_instances.append(instances.text)
    except NoSuchElementException:
        No_of_instances.append('Not Present')

    try:
        no_of_attr = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p')
        No_of_attribute.append(no_of_attr.text)
    except NoSuchElementException:
        No_of_attribute.append('Not Present')

    try:
        yr = driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        Year.append(yr.text)
    except NoSuchElementException:
        Year.append('Not Present')
```

```python
len(Dataset_name),len(Data_type),len(Task),len(Attribute_type),len(No_of_instances),len(No_of_attribute),len(Year)
```

```python
df = pd.DataFrame({'Dataset_name':Dataset_name,'Data_type':Data_type,'Task':Task,'Attribute_type':Attribute_type,'No_of_instances':No_of_instances,'No_of_attribute':No_of_attribute,'Year':Year})
df
```
