```python
from bs4 import BeautifulSoup
import requests
```

5) Please visit https://www.cnbc.com/world/?region=world and scrap-
 a) headings
b) date
c) News link

```python
url_ = 'https://www.cnbc.com/2024/06/14/europe-stocks-head-for-higher-open-to-round-off-choppy-week.html'

news = requests.get(url_)
news
```

```python
soup2 = BeautifulSoup(news.content)
soup2
```

```python
headings = soup2.find('h1',class_="ArticleHeader-headline")
headings
```

```python
headings.text
```

```python
date = soup2.find('time',datetime="2024-06-14T06:14:45+0000")
date
```

```python
date.text
```

```python
news_link = soup2.find('a',class_="WatchLive-container")
news_link
```

```python
news_link.text
```

 6) Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloadedarticles/ and scrapa) Paper title
 b) date
c) Author

```python
url= 'https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/'

page1 = requests.get(url)
page1
```

```python
soup3 = BeautifulSoup(page1.content)
soup3
```

```python
title = soup3.find('h1')
title
```

```python
print('Title :-',title.text)
```

```python
author = soup3.find('span',class_="lead mb-3 d-block")
replace=author.text.replace('\r\n                                                    ','')
result = replace.replace('\r\n                                                ','')
print('Author :-',result)
```
