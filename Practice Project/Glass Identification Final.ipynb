```python
#Divide the classes here into 2 categories - window glass (classes 1-4)  as 0 and non-window glass (classes 5-7) as 1
df['Type of glass']=[1 if x>=5 else 0 for x in df['Type of glass']]
df['Type of glass'].value_counts()
```

Separate the Variables

```python
x = df.drop('Type of glass',axis = 1)
y = df['Type of glass']
```

```python
x.shape
```

```python
y.shape
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['VIF values'] = [ variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif['features'] = x.columns
vif
```

Here we see that the many features have high vif value, above the threshold value we should drop those columns but instead of drop those columns i am using PCA

After Dividing the classes here into 2 categories - window glass (classes 1-4)  as 0 and non-window glass (classes 5-7) as 1, using oversampling to avoid duplicate data and create new data here to balance the minority and mejority classes. Using SMOTE to increase the minority class.

#### Scalling the Features

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale
```

```python
x = scale.fit_transform(x)
x
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
smote
```

```python
x1,y1 = smote.fit_resample(x,y)
```

```python
y1.value_counts()
```

Handling Multicollinearity by using PCA

Here we see that the many features have high vif value, above the threshold value we should drop those columns but instead of drop those columns i am using PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
pca
```

```python
x = pca.fit_transform(x)
x.shape
```

```python
pd.DataFrame(data = x)
```

### Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.20, random_state = 42)
```

```python
LR = LogisticRegression()
LR
```

```python
DTC = DecisionTreeClassifier()
DTC
```

```python
DTC.fit(x_train,y_train)
DTC_pred = DTC.predict(x_test)
print(accuracy_score(y_test,DTC_pred))
print(classification_report(y_test,DTC_pred))
print(confusion_matrix(y_test,DTC_pred))
```

```python
LR.fit(x_train,y_train)
LR_pred = LR.predict(x_test)
print(accuracy_score(y_test,LR_pred))
print(classification_report(y_test,LR_pred))
print(confusion_matrix(y_test,LR_pred))
```

```python
RF = RandomForestClassifier()
RF
```

```python
RF.fit(x_train,y_train)
RF_pred = RF.predict(x_test)
print(accuracy_score(y_test,RF_pred))
print(classification_report(y_test,RF_pred))
print(confusion_matrix(y_test,RF_pred))
```

```python
ET = ExtraTreesClassifier()
ET
```

```python
ET.fit(x_train,y_train)
ET_pred = ET.predict(x_test)
print(accuracy_score(y_test,ET_pred))
print(classification_report(y_test,ET_pred))
print(confusion_matrix(y_test,ET_pred))
```

```python
GB = GradientBoostingClassifier()
GB
```

```python
GB.fit(x_train,y_train)
GB_pred = GB.predict(x_test)
print(accuracy_score(y_test,GB_pred))
print(classification_report(y_test,GB_pred))
print(confusion_matrix(y_test,GB_pred))
```

```python
AD = AdaBoostClassifier()
AD
```

```python
AD.fit(x_train,y_train)
AD_pred = AD.predict(x_test)
print(accuracy_score(y_test,AD_pred))
print(classification_report(y_test,AD_pred))
print(confusion_matrix(y_test,AD_pred))
```

```python
BG = BaggingClassifier()
BG
```

```python
BG.fit(x_train,y_train)
BG_pred = BG.predict(x_test)
print(accuracy_score(y_test,BG_pred))
print(classification_report(y_test,BG_pred))
print(confusion_matrix(y_test,BG_pred))
```

```python
svc = SVC()
```

```python
svc.fit(x_train,y_train)
svc_pred = svc.predict(x_test)
print(accuracy_score(y_test,svc_pred))
print(classification_report(y_test,svc_pred))
print(confusion_matrix(y_test,svc_pred))
```

### HyperParameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_param = {
    'criterion' : ['friedman_mse', 'squared_error'],
    'n_estimators' :[20,50,100,150],
    'learning_rate' :[0.01,0.1,0.2],
    'max_depth' :[3,5,7,10],
    'random_state':[42,150]
    }

grid_search = GridSearchCV(estimator = GB,param_grid=grid_param,scoring ='accuracy',cv=5)
grid_search.fit(x_train,y_train)

best_param = grid_search.best_params_
print(best_param)

best_result = grid_search.best_score_
print(best_result)

print(round(best_result,2))
```

```python
GB_model = GradientBoostingClassifier(criterion= 'friedman_mse', max_depth=3,n_estimators= 150,random_state= 42)
GB_model.fit(x_train,y_train)
pred = GB_model.predict(x_test)
accuracy = accuracy_score(y_test,pred)
conf_matrics = confusion_matrix(y_test,pred)
Class_report = classification_report(y_test,pred)
```

```python
print('accuracy_score  :',(accuracy))
print('confusion_matrix  :',(conf_matrics))
print('classification_report  :',(Class_report))
```

```python
grid_param = {
    'criterion' : ['gini','entropy'],
    'n_estimators' :[20,50,100,150],
    'max_features' :['log2','sqrt'],
    'max_depth' :[3,5,7,10],
    'random_state':[42,150]
    }

grid_search = GridSearchCV(estimator = RF,param_grid=grid_param,scoring ='accuracy',cv=5)
grid_search.fit(x_train,y_train)

best_param = grid_search.best_params_
print(best_param)

best_result = grid_search.best_score_
print(best_result)

print(round(best_result,2))
```
