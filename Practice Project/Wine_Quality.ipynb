# Separate the Variables

```python
x=data.drop("quality",axis=1)
y=data["quality"]
```

# Using Scalling Method

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x= pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

# Finding Variance Inflation Factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

VIF = pd.DataFrame()
VIF["vif values"] = [ variance_inflation_factor(x.values,i)for i in range(len(x.columns))]
VIF["features"] = x.columns
VIF
```

find multicollinearity problem by using VIF .... here fixed acidity and density are correlation with other feature

here we drop those feature

```python
x.drop("fixed acidity",axis = 1,inplace=True)
```

```python
x.drop("density",axis = 1,inplace=True)
```

```python
VIF = pd.DataFrame()
VIF["vif values"] = [ variance_inflation_factor(x.values,i)for i in range(len(x.columns))]
VIF["features"] = x.columns
VIF
```
